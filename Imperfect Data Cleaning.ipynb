## Time for the questions!
I feel like I'm about to do some cop out nonsense but the most logical questions are how many parks are there with specific features. The following is a function I got from Stack Overflow from the user [ballsatballsdotballs](https://stackoverflow.com/a/29546836/5741205) to get a rough estimation of the distance of parks from Code Fellows:

```python
def get_distance(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    km = 6367 * c
    return km
```

### Question 1
How far are all the parks with boat launches from Code Fellows:

```python
distance, boats_idx = [], []
cf_lon, cf_lat = -122.3519, 47.6182
for field in park_data:
    for idx, data in enumerate(park_data[field]):
        if data == 'Boat Launches':
            latitude = park_data.iloc[idx].Latitude
            longitude = park_data.iloc[idx].Longitude
            distance.append(get_distance(longitude, latitude, cf_lon, cf_lat))
            boats_idx.append(idx)
```

```python
plt.scatter(boats_idx, distance, c='blue')
plt.xticks([])
plt.xlabel('Code Fellows')
plt.ylabel('Distance from Code Fellows in miles')
plt.title('Distance of parks with boat launches from Code Fellows')
plt.show()
```

### Question 2
How far are all the parks with gardens from Alki beach:

```python
distance, gardens_idx = [], []
alki_lon, alki_lat = -122.4114, 47.5791
for field in park_data:
    for idx, data in enumerate(park_data[field]):
        if data == 'Gardens':
            lat = park_data.iloc[idx].Latitude
            lon = park_data.iloc[idx].Longitude
            distance.append(get_distance(lon, lat, alki_lon, alki_lat))
            gardens_idx.append(idx)
```

### Question 5
How far are parks with both learning centers and community centers from both Code Fellows and Alki:

```python
cf_distance = []
alki_distance = []
community_idx = []
learning_idx = []
cf_learning_dist = []
alki_learning_dist = []
for field in park_data:
    for idx, data in enumerate(park_data[field]):
        if data == 'Environmental Learning  Centers':
            lat = park_data.iloc[idx].Latitude
            lon = park_data.iloc[idx].Longitude
            cf_learning_dist.append(get_distance(lon, lat, cf_lon, cf_lat))
            alki_learning_dist.append(get_distance(lon, lat, alki_lon, alki_lat))
            learning_idx.append(idx)
        elif data == 'Community Centers':
            lat = park_data.iloc[idx].Latitude
            lon = park_data.iloc[idx].Longitude
            cf_distance.append(get_distance(lon, lat, cf_lon, cf_lat))
            alki_distance.append(get_distance(lon, lat, alki_lon, alki_lat))
            community_idx.append(idx)
```

```python
plt.subplot(2, 2, 1)
plt.scatter(community_idx, cf_distance)
plt.title('Distance of parks with community centers from Code Fellows')
plt.xticks([])
plt.xlabel('Code Fellows')
plt.ylabel('Distance from Code Fellows in miles')

plt.subplot(2, 2, 2)
plt.scatter(community_idx, alki_distance)
plt.title('Distance of parks with community centers from Alki')
plt.xticks([])
plt.xlabel('Alki')
plt.ylabel('Distance from Alki in miles')

plt.subplot(2, 2, 3)
plt.scatter(learning_idx, cf_learning_dist)
plt.title('Distance of parks with learning centers from Code Fellows')
plt.xticks([])
plt.xlabel('Code Fellows')
plt.ylabel('Distance from Code Fellows in miles')

plt.subplot(2, 2, 4)
plt.scatter(learning_idx, alki_learning_dist)
plt.title('Distance of parks with learning centers from Alki')
plt.xticks([])
plt.xlabel('Alki')
plt.ylabel('Distance from Alki in miles')
plt.subplots_adjust(top=1.5, right=2)
plt.show()
```
