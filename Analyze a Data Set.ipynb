```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
boston = datasets.load_boston()
```

## The dataset
This dataset is about Boston housing prices and crime rates

```python
print(boston.DESCR)
```

```python
river_crime, no_river_crime, river_student, no_river_student = [], [], [], []
for item in boston.data:
    if item[3] == 1:
        river_crime.append(item[0])
        river_student.append(item[10])
    else:
        no_river_crime.append(item[0])
        no_river_student.append(item[10])
plt.scatter(sorted(no_river_student, key=int), sorted(no_river_crime, key=int), s=30, c="red")
plt.scatter(sorted(river_student, key=int), sorted(river_crime, key=int), s=30, c="blue")
plt.ylabel("Crime per Capita")
plt.xlabel("Pupil-to-Teacher Ratio")
plt.title("Crime/Pupil-To-Teacher Ratio in the Boston Suburbs")
plt.show()
```

```python
number_of_black_residents, employment_centers = [], []
for item in boston.data:
    number_of_black_residents.append(item[7])
    employment_centers.append(item[11])
plt.scatter(sorted(number_of_black_residents, key=int), sorted(employment_centers, key=int), s=5, c='blue')
plt.ylabel('# of Black Residents')
plt.xlabel('Distance to Employment Centers in Miles')
plt.title('Number of Black residents/Distance to Employment Centers')
plt.show()
```
